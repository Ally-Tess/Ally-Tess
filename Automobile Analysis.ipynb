```r
# Model 1

op=par(mfcol=c(2,2))
plot(highwaympg curbweight + price+ horsepower , data=auto)
m1<-lm(highwaympg curbweight + price+horsepower , data=auto)
plot(m1)
abline(m,col="red")
summary(m1)
op=par(mfcol=c(2,2))
par(op)
anova(m1)  
```

```r
# Model without interaction term  

op=par(mfcol=c(2,2))
m1<-lm(highwaympg aspiration + curbweight, data=auto)
plot(m1)
abline(m,col="red")
summary(m1)
op=par(mfcol=c(2,2))
par(op)
anova(m1)
```

```r
# With interaction term

op=par(mfcol=c(2,2))
m1<-lm(highwaympg aspiration*curbweight, data=auto)
plot(m1)
abline(m,col="red")
summary(m1)
op=par(mfcol=c(2,2))
par(op)
anova(m1)
```
